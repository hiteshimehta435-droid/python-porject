```python
# fundamental Booster porject
print("Welcome to the Interactive personal Data collecter!")
# input from user
name = input("Enter your Name: ")
age = input("Enter your Age: ")
height = float(input("Enter your height (in feet): "))
fav_num = int(input("Enter your Favourite namber: "))

print("\n------------ USER INFORMATION -----------------")

# Birth year calculation
birth_year = 2026 - int(age)

# Display information
print("Name :", name)
print("Age :", age)
print("Height :", height)
print("Favourite Number :", fav_num)
print("Birth year :", birth_year)

# Arithmetic Operstions
print("\n----------- ARITHMETIC OPERATIONS -------------")

addition = int(age) + fav_num
subtraction = fav_num - int(age)
multipliction = int(age) * fav_num
division = fav_num / int(age)

print("Age + Favourite Number =", addition)
print("Favourite Number - Age =", subtraction)
print("Age * Favourite Number =", multipliction)
print("Favourite Number / Age =", division)

# Data Types
print("\n---------- DATA TYPES ------------------------")

print("Type of Name =", type(name))
print("Type of Age =", type(age))
print("Type of Height =", type(height))
print("Type of Favourite Number =", type(fav_num))

# Memory Address
print("\n--------- MEMORY ADDRESS ---------------------")

print("ID of Name =", id(name))
print("ID of Age =", id(age))
print("ID of height =", id(height))
print("ID of Favourite Number =", id(fav_num))

# Type Conversion
print("\n---------- Type conversion ------------------")

age_float = float(age)
fav_str = str(fav_num)
print("Age converted to Float =", age_float)
print("favourite Number converted to string =", fav_str)

print("\nThank you for using personal Data collector!")
```
